```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib nbagg
import random
ran_Ax = [0]*30
ran_Ay= [0]*30
ran_Bx= [0]*30
ran_By= [0]*30
for i in range(0,30,1):
    ran_Ax[i]= random.randint(1,10)
    ran_Ay[i]= random.randint(10,20)
    ran_Bx[i]= random.randint(10,20)
    ran_By[i]= random.randint(0,10) 

 
F_point = [8,9]
 
plt.figure()
plt.scatter(ran_Ax,ran_Ay)
plt.scatter(ran_Bx,ran_By)
plt.scatter(F_point[0],F_point[1], marker='*')
 
plt.show()


def k_neighbor(f_point,A_x,A_y,B_x,B_y,K):
    try:
   #try:
        a=list(np.zeros(len(A_y)))
        b=list(np.zeros(len(B_y)))
        a_f_dist=list(np.zeros(len(A_y)))
        b_f_dist=list(np.zeros(len(B_y)))

        for i in range(0,len(A_y),1):
            a[i]=([A_x[i],A_y[i]])
            b[i]=([B_x[i],B_y[i]])
            a_f_dist[i]=np.sqrt(((a[i][0]-f_point[0])**2)+((a[i][1]-f_point[1])**2))
            b_f_dist[i]=np.sqrt(((b[i][0]-f_point[0])**2)+((b[i][1]-f_point[1])**2))
        a_dist=a_f_dist.copy()
        b_dist=b_f_dist.copy()
        a_dist.sort()
        b_dist.sort()
        all_dist=a_f_dist+b_f_dist
        all_dist.sort()
        #n=int(input("num"))
        #a_cnt=[]
        for i in range (0,len(all_dist[:K]),1):
            a_cnt=list(a_f_dist<=all_dist[i]).count(True)
            b_cnt=list(b_f_dist<=all_dist[i]).count(True)
            a_bool_list= list(a_f_dist<=all_dist[i])
            b_bool_list= list(b_f_dist<=all_dist[i])
            a_index_num=[ r for r, x in enumerate( a_bool_list ) if x == True ]
            b_index_num=[ q for q, x in enumerate( b_bool_list ) if x == True ]
        for h in a_index_num:
            #a[h]
            print("A그룹 point ="+str(a[h]))
        for q in b_index_num:
            #a[h]
            print("B그룹 point ="+str(b[q]))

            
        #res1=print("K의 갯수= "+str(K)+"\nA그룹 갯수 = "+str(a_cnt) +"\nB그룹 갯수 = "+str(b_cnt))
        res1=print("\nK의 갯수= "+str(K)+"\nA그룹 갯수 = "+str(a_cnt) +"\nA그룹 distance ="+str(a_dist[:a_cnt]) +"\nB그룹 갯수 = "+str(b_cnt)+"\nB그룹 distance ="+str(b_dist[:b_cnt]))
        if a_cnt > b_cnt :
            res2=print("F_point가 속한 그룹은 A")
        elif a_cnt < b_cnt :
            res2=print("F_point가 속한 그룹은 B")

        elif a_cnt==b_cnt:
            if sum(a_f_dist[:int(K/2)]) < sum(b_f_dist[:int(K/2)]):
                res3=print("\n주간적인 견해의 F_point가 속한 그룹은 A(거리의 합을 기준으로...),\n하지만 A그룹에 포함된 점의 갯수와 B그룹에 포함된 점의 갯수가 같기 때문에\n신뢰도 매우 낮음")
            elif sum(a_f_dist[:int(K/2)]) > sum(b_f_dist[:int(K/2)]):    
                res3=print("\n주간적인 견해의 F_point가 속한 그룹은 B(거리의 합을 기준으로...),\n하지만 A그룹에 포함된 점의 갯수와 B그룹에 포함된 점의 갯수가 같기 때문에\n신뢰도 매우 낮음")
            else:
                res3=print("\n예측불가")
    except UnboundLocalError :
        pass
    
        return res1,res2,res3    



k_neighbor(F_point,ran_Ax,ran_Ay,ran_Bx,ran_By,1)
```
